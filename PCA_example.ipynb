```
import numpy as np
import pandas as pd
from IPython.display import display
```

```
def generate_image_array(size=28, sparsity=5, flatten=False): #just demo, not real image
  image_array = (np.random.randint(low=0, high=255, size=size**2)*np.random.choice([1]+[0]*sparsity,size=size**2)).reshape(size,size)
  if flatten:
    image_array = image_array.flatten()
  return image_array

image_array = generate_image_array()
print(image_array.shape)
# image_array
```

```
image_array_flattened = generate_image_array(flatten=True)
print(image_array_flattened.shape)
# image_array_flattened
```

```
df_arrays = pd.DataFrame(index=np.arange(1006), columns=["pixel_%i"%i for i in range(28**2)])
df_arrays.loc[:] = [generate_image_array(flatten=True) for _ in range(1006)]
df_arrays.index.name="fontID"
df_arrays
```

```
from sklearn.decomposition import PCA, KernelPCA
pca = PCA(n_components=50)
df_arrays_reduced = pca.fit_transform(df_arrays)
df_arrays_reduced = pd.DataFrame(df_arrays_reduced, columns=["PC_%i"%i for i in range(df_arrays_reduced.shape[1])], index=df_arrays.index)
print("ðŸ‘‰ Explained variance: %.2f%%"%(pca.explained_variance_ratio_.sum()*100))
df_arrays_reduced
```
